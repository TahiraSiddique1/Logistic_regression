```python
 #oad the dataset
import matplotlib.pyplot as  plt
df = pd.read_csv(r'banking.csv')

# Plot y for counts
df['y'].value_counts().plot(kind='bar')
plt.title('Counts of y')
plt.xlabel('y')
plt.ylabel('Count')
plt.show()

# Plot y for job
df.groupby(['job', 'y'])['y'].count().unstack().plot(kind='bar', stacked=True)
plt.title('Counts of y by job')
plt.xlabel('Job')
plt.ylabel('Count')
plt.show()

# Plot y for martial
df.groupby(['marital', 'y'])['y'].count().unstack().plot(kind='bar', stacked=True)
plt.title('Counts of y by marital')
plt.xlabel('Marital')
plt.ylabel('Count')
plt.show()

```

```python
# Plot y for loan
df.groupby(['loan', 'y'])['y'].count().unstack().plot(kind='bar', stacked=True)
plt.title('Counts of y by loan')
plt.xlabel('Loan')
plt.ylabel('Count')

plt.show()
```

```python
# Plot y for poutcome
df.groupby(['poutcome', 'y'])['y'].count().unstack().plot(kind='bar', stacked=True)
plt.title('Counts of y by poutcome')
plt.xlabel('Poutcome')
plt.ylabel('Count')
plt.show()

```
